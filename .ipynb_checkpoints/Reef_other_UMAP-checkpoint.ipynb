```python
import numba
from numba import jit
import scipy

MIN_OVERLAP = 0.9
N_RANDOM_SHUFFLE = 10

@jit(nopython=True)
def calc_zscore(spec):
    mn = np.mean(spec)
    std = np.std(spec)
    for i in range(spec.shape[0]):
        for j in range(spec.shape[1]):
            spec[i,j] = (spec[i,j]-mn)/std
    return spec

@jit(nopython=True)
def preprocess_spec_numba(spec):
    spec = calc_zscore(spec)
    spec = np.where(spec > 3, 3, spec)
    spec = np.where(spec < 0, 0, spec)   
    return spec

def pad_spectro_bothsides(spec,maxlen):
    full_padding = maxlen - spec.shape[1]
    padding_left = int(full_padding/2)
    padding_right = full_padding - padding_left
    z_left = np.zeros((spec.shape[0],padding_left))
    z_right = np.zeros((spec.shape[0],padding_right))
    padded_spec=np.append(z_left, spec, axis=1)
    padded_spec=np.append(padded_spec, z_right, axis=1)
    return padded_spec
```

```python
@jit(nopython=True)
def spec_dist_numba(s_1, s_2):   
    dist = np.sum((np.subtract(s_1, s_2)*np.subtract(s_1, s_2)))  
    # Normalize to random shuffling
    random_dist = calc_random_dist_numba(s_1,s_2)
    norm_dist = dist / random_dist   
    return norm_dist

@jit(nopython=True)
def calc_random_dist_numba(s_1, s_2):
    dists = np.full((N_RANDOM_SHUFFLE),3.)
    s_1_shuffled = np.copy(s_1)
    for i in range(N_RANDOM_SHUFFLE):
        np.random.shuffle(s_1_shuffled)
        dists[i] = np.sum(np.subtract(s_1_shuffled, s_2)*np.subtract(s_1_shuffled, s_2)) 
    return(np.mean(dists))   
```

```python
@jit(nopython=True)
def calc_shiftdist(spec_s, spec_l):
    
    len_s = spec_s.shape[1]
    len_l = spec_l.shape[1]

    nfreq = spec_s.shape[0]
    

    # define start position
    min_overlap_frames = int(MIN_OVERLAP * len_s)
    start_timeline = min_overlap_frames-len_s
    max_timeline = len_l - min_overlap_frames
    
    n_of_calculations = (max_timeline+1-start_timeline)+(max_timeline+1-start_timeline)

    distances = np.full((n_of_calculations),3.)
    spec_s_configs = []
    spec_l_configs = []

    count=0
    
    for timeline_p in range(start_timeline, max_timeline+1):
        #print("timeline: ", timeline_p)
        # mismatch on left side
        if timeline_p < 0:
            #print("mismatch on left")

            len_overlap = len_s - abs(timeline_p)
            
            pad_s = np.full((nfreq, (len_l-len_overlap)),0.)
            pad_l = np.full((nfreq, (len_s-len_overlap)),0.)

            s_config = np.append(spec_s, pad_s, axis=1)
            l_config = np.append(pad_l, spec_l, axis=1)

        # mismatch on right side
        elif timeline_p > (len_l-len_s):
            #print("mismatch on right")
            
            len_overlap = len_l - timeline_p

            pad_s = np.full((nfreq, (len_l-len_overlap)),0.)
            pad_l = np.full((nfreq, (len_s-len_overlap)),0.)

            s_config = np.append(pad_s, spec_s, axis=1)
            l_config = np.append(spec_l, pad_l, axis=1)

        # no mismatch on either side
        else:
            #print("no mismatch")
            len_overlap = len_s
            start_col_l = timeline_p
            end_col_l = start_col_l + len_overlap

            pad_s_left = np.full((nfreq, start_col_l),0.)
            pad_s_right = np.full((nfreq, (len_l - end_col_l)),0.)
            #print("left +", start_col_l, ", right +", (len_l - end_col_l))
            l_config = spec_l
            s_config = np.append(pad_s_left, spec_s, axis=1)
            s_config = np.append(s_config, pad_s_right, axis=1)
        
        size = s_config.shape[0]*s_config.shape[1]
        #print("s: ", s_config.shape)
        #print("l: ", l_config.shape)

        distances[count] = (np.sum(np.subtract(s_config, l_config)*np.subtract(s_config, l_config))) / size
        
        spec_s_configs.append(s_config)
        spec_l_configs.append(l_config)
                            
        count = count + 1

            
    ind = np.argmin(distances)
    min_dist = distances[ind]

    best_config_s = spec_s_configs[ind]
    best_config_l = spec_l_configs[ind]

    dist = spec_dist_numba(best_config_s,best_config_l)

    #return ind, dist, best_config_s, best_config_l
    return dist
```

```python
s_list = [preprocess_spec_numba(np.abs(s)) for s in spec_df.ampli_specs]
```

```python
import timeit
import random
```

```python
sample_list = random.sample(s_list,30)
time_diffs = []

distmat = np.zeros((len(sample_list), len(sample_list)))
n_comps = ((len(sample_list)**2) - len(sample_list)) / 2

count=0
for i in range(len(sample_list)):
    for j in range(i+1,len(sample_list)):
    
    #print(i, j)
        print('\rProcessing ',count, end='')

        starttime = timeit.default_timer()

        specs = [sample_list[i], sample_list[j]]
        spec_ls = [s.shape[1] for s in specs]
        longer = np.argmax(spec_ls)
        len_longer = spec_ls[longer]
        len_shorter = spec_ls[int(not longer)]

        longer_spec = specs[longer]
        shorter_spec = specs[int(not longer)]

        if len_longer>=(2*len_shorter):
            padded_spec = pad_spectro_bothsides(shorter_spec, len_longer)
            distmat[i,j] = spec_dist_numba(padded_spec,longer_spec)
        else:
            distmat[i,j] = calc_shiftdist(shorter_spec, longer_spec)
    
    time_diff = timeit.default_timer() - starttime
    time_diffs.append(time_diff)
    count = count +1
```
